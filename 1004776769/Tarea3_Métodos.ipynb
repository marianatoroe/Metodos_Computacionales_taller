```python
paises = ["Argentina","Bolivia","Chile","Colombia","Costa Rica","Ecuador","El Salvador","Guatemala","Honduras","Mexico","Panama",
                       "Paraguay","Peru","Dominican Republic","Uruguay","Venezuela"] #comparación de países latinoamericanos
datos = df3[["total_vaccinations","people_vaccinated","people_fully_vaccinated","daily_vaccinations_raw","daily_vaccinations",
                         "total_vaccinations_per_hundred","people_vaccinated_per_hundred","people_fully_vaccinated_per_hundred",
                         "daily_vaccinations_per_million"]]

lista1=[] #listas para cada conjunto de datos
lista2=[]
lista3=[]
lista4=[]
lista5=[]
lista6=[]
lista7=[]
lista8=[]
lista9=[]

#asignación de los valores iniciales para completar las listas 
for i in paises: 
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["total_vaccinations"].max()
  lista1.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["people_vaccinated"].max()
  lista2.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["people_fully_vaccinated"].max()
  lista3.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["daily_vaccinations_raw"].mean()
  lista4.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["daily_vaccinations"].mean()
  lista5.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["total_vaccinations_per_hundred"].mean()
  lista6.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["people_vaccinated_per_hundred"].mean()
  lista7.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["people_fully_vaccinated_per_hundred"].mean()
  lista8.append(num)

for i in paises:
  df = df3.loc[df3.loc[:, 'country'] == i]
  num = df["daily_vaccinations_per_million"].mean()
  lista9.append(num)

dic=[] #creación del DataFrame con la información de los países latinoamericanos
dataf=[]
for i in range(0,16,1): #se itera sobre el número total de países latinoamericanos
  dic={"total_vaccinations":lista1[i], "people_vaccinated":lista2[i], "people_fully_vaccinated":lista3[i], "daily_vaccinations_raw":lista4[i],
       "daily_vaccinations":lista5[i], "total_vaccinations_per_hundred":lista6[i], "people_vaccinated_per_hundred":lista7[i],
       "people_fully_vaccinated_per_hundred":lista8[i], "daily_vaccinations_per_million":lista9[i]}
  dataf.append(dic)
dataff = pd.DataFrame(dataf)
dataff.index = paises
print(dataff)

dataffintervalo1 = dataff[dataff["total_vaccinations"]>=(1.5*10**6)] #número de vacunados en el intervalo > 1.5E6
dataffintervalo2 = dataffintervalo1[dataffintervalo1["total_vaccinations"]<=(2*10**6)] #número de vacunados en el intervalo < 2.0E6
print("Los pais/paises con el numero de vacunados en el intervalo [1.5E6, 2.0E6] es/son:")
print(dataffintervalo2)
print("Ningún país latinoamericano cumple con la condición") #conclusión
```
